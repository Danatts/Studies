```python
# Importamos la base de datos
df = sns.load_dataset('iris')
df.head()
```

```python
# Reemplazamos datos
df['target'] = 0
df.loc[df['species'] == 'versicolor', 'target'] = 1
df.loc[df['species'] == 'virginica', 'target'] = 2
```

```python
# Datos ordenados

# Con ordean descendente y ascendente
# df_sorted = df.sort_values(['petal_length', 'petal_width'], ascending = [False, True])

df_sorted = df.sort_values(['petal_length', 'petal_width'])

plt.plot(df_sorted.petal_length, df_sorted.petal_width)
plt.xlabel('Largo (cm)')
plt.ylabel('Ancho (cm)')
plt.title('Medidas del pétalo')
plt.savefig(fname='line.png')
# plt.close()
```
