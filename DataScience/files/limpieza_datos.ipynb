### Carga de datos usando la funciÃ³n `open`

A la hora de trabajar con datos muy grandes, pandas puede ser menos indicado ya que carga toda la informaciÃ³n en la memoria RAM y puede acaparar gran parte de esta. Para estos casos se puede usar la funciÃ³n nativa de python `open`.

```python
df = open('../Datasets/titanic/titanic.csv', 'r')
```

```python
cols = df.readline().split(',') # Nombre de las columnas
```

```python
# Creamos un diccionario con cada una de las columnas como llaves
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
# Agregamos los datos en forma de lista a las llaves del diccionario
counter = 0
for line in df:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1

print(f'El data set tiene {counter} filas y {len(cols)} columnas')
```

## Tratar con valores nulos

### Conocer los datos nulos

```python
# Devuelve un booleano indicando quÃ© valores son nulos
pd.isnull(df)
```

```python
# NÃºmero de valores nulos por columna
pd.isnull(df).sum()
```

### Borrar valores faltantes

```python
# Axis 0: borra las filas, Axis 1 = borra las columnas
# Opciones how: all (todos las columnas son NA), any (alguna columna tiene datos NA)
df.dropna(axis = 0, how = 'any')
```

### CÃ³mputo de los valores faltantes

```python
# Reemplazar todos los NA por un valor determinado
df.fillna(0)
df['body'].fillna(0)
df['age'].fillna(df['age'].mean())
```

```python
# Reemplazar los NA con el dato anterior o posterior
df.fillna(method = 'bfill') # Sustituir por anterior
df.fillna(method = 'ffill') # Sustituir por posterior
```

## Variable Dummy

Son variables separadas para cada categorÃ­a que se crean para las variables categÃ³ricas.

```python
dummy_sex = pd.get_dummies(df['sex'], prefix = 'sex')
dummy_sex
```

Una vez se crean las variables dummies se puede:
- AÃ±adir las columnas dummies al dataset original
- Reemplazar la variable original por las dummies

```python
df_dummy = df.drop(['sex'], axis = 1)
pd.concat([df_dummy, dummy_sex], axis = 1)
```
