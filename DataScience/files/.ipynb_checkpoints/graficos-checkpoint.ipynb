```python
# Importamos la base de datos
df = sns.load_dataset('iris')
df.head()
```

```python
# Reemplazamos datos
df['target'] = 0
df.loc[df['species'] == 'versicolor', 'target'] = 1
df.loc[df['species'] == 'virginica', 'target'] = 2
```

## Gráfico de línea

```python
# Datos ordenados

# Con ordean descendente y ascendente
# df_sorted = df.sort_values(['petal_length', 'petal_width'], ascending = [False, True])

df_sorted = df.sort_values(['petal_length', 'petal_width'])

plt.plot(df_sorted.petal_length, df_sorted.petal_width)
plt.xlabel('Largo (cm)')
plt.ylabel('Ancho (cm)')
plt.title('Medidas del pétalo')
plt.savefig(fname='line.png')
# plt.close()
```

```python
plt.plot(df_sorted.loc[df_sorted.species == 'setosa', 'petal_length'], df_sorted.loc[df_sorted.species == 'setosa', 'petal_width'])
plt.title('Médidas de pétalo (Setosa)')
plt.xlabel('Largo (cm)')
plt.xlabel('Ancho (cm)')
plt.close()
```

```python
plt.plot(df_sorted.loc[df_sorted.species == 'virginica', 'petal_length'], df_sorted.loc[df_sorted.species == 'virginica', 'petal_width'])
plt.title('Médidas de pétalo (Virginica)')
plt.xlabel('Largo (cm)')
plt.xlabel('Ancho (cm)')
plt.close()
```

```python
plt.plot(df_sorted.loc[df_sorted.species == 'versicolor', 'petal_length'], df_sorted.loc[df_sorted.species == 'versicolor', 'petal_width'])
plt.title('Médidas de pétalo (Versicolor)')
plt.xlabel('Largo (cm)')
plt.xlabel('Ancho (cm)')
plt.close()
```

```python
plt.plot(df_sorted.loc[df_sorted.species == 'setosa', 'petal_length'], df_sorted.loc[df_sorted.species == 'setosa', 'petal_width'], 'b.-', label='setosa')
plt.plot(df_sorted.loc[df_sorted.species == 'versicolor', 'petal_length'], df_sorted.loc[df_sorted.species == 'versicolor', 'petal_width'], 'ro--', label='versicolor')
plt.plot(df_sorted.loc[df_sorted.species == 'virginica', 'petal_length'], df_sorted.loc[df_sorted.species == 'virginica', 'petal_width'], 'g^-.', label='virginica')
plt.title('Medidas del pétalo por especies')
plt.xlabel('Ancho (cm)')
plt.ylabel('Largo (cm)')
plt.legend()
plt.show()
plt.close()
```

## Gráficos de dispersión

```python
plt.clf()

plt.scatter(df.petal_length, df.petal_width)
plt.title('Medidas del pétalo')
plt.xlabel('Ancho (cm)')
plt.ylabel('Largo (cm)')
plt.show()
plt.close()
```

```python
colors = ['blue', 'red', 'green']
for val in df.target.unique():
    print(val, colors[val])
```

```python
for val in df['target'].unique():
    plt.scatter(df.loc[df['target'] == val, 'petal_length'],
                df.loc[df['target'] == val, 'petal_width'],
                color = colors[val],
                label = df.species.unique()[val])
plt.title('Medidas del pétalo por especie')
plt.xlabel('Ancho (cm)')
plt.ylabel('Largo (cm)')
plt.legend(loc = 'lower right')
plt.show()
plt.close()
```

```python
x = [1, 2, 3, 4, 5]
y = [4, 3, 2, 5, 3]

plt.scatter(x, y)
for index in range(len(x)):
    plt.annotate(f'Punto {index + 1}', (x[index], y[index]))
```

## Gráficos de histogramas

```python
plt.hist(df.petal_length, bins = 50) # Con el parámetro 'bins' se ajusta el número de muestras por categoría
```

```python
for val in df.target.unique():
    plt.hist(df.loc[df.target == val, 'petal_length'],
    color = colors[val],
    label = df.species.unique()[val],
    alpha = 0.5)
plt.title('Medidas del pétalo por especie')
plt.xlabel('Largo (cm)')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
plt.close()
```

## Gráficos de barras

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
plt.bar(titanic.sex.sort_values().unique(), titanic.groupby('sex').survived.count())
```

```python
plt.bar(titanic['class'].sort_values().unique(), titanic.groupby('class').survived.count(),
       color = ['cyan', 'yellow', 'violet'])
```

## Gráfico de torta

```python
class_counts = titanic['class'].value_counts()
plt.pie(class_counts,
       colors = ['cyan', 'yellow', 'violet'],
       labels = class_counts.index,
       autopct = '%0.2f%%')
```
